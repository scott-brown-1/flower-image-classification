```python
import pandas as pd
import numpy as np
```

```python
scott_probas = pd.read_csv('./y_pred_4.csv')
ty_probas = pd.read_csv('./probabilities.csv')  
```

```python
#daisy	dandelion	rose	sunflower	tulip
daisy = []
dandelion = []
rose = []
sunflower = []
tulip = []

scott_probas_clean = pd.DataFrame({'ID': scott_probas['ID']})
for row in range(len(scott_probas)):
    proba_array = np.array(scott_probas['Prediction'][row].replace("'","").replace('[','').replace(']','').split(',')).astype(float)
    daisy.append(proba_array[0])
    dandelion.append(proba_array[1])
    rose.append(proba_array[2])
    sunflower.append(proba_array[3])
    tulip.append(proba_array[4])

scott_probas_clean['daisy'] = daisy
scott_probas_clean['dandelion'] = dandelion
scott_probas_clean['rose'] = rose
scott_probas_clean['sunflower'] = sunflower
scott_probas_clean['tulip'] = tulip
```

```python
## Average the probabilities of both dataframe
final_probas = pd.DataFrame({'ID': scott_probas_clean['ID']})
# final_probas['daisy'] = (scott_probas_clean['daisy'] + ty_probas['daisy'])/2
# final_probas['dandelion'] = (scott_probas_clean['dandelion'] + ty_probas['dandelion'])/2
# final_probas['rose'] = (scott_probas_clean['rose'] + ty_probas['rose'])/2
# final_probas['sunflower'] = (scott_probas_clean['sunflower'] + ty_probas['sunflower'])/2
# final_probas['tulip'] = (scott_probas_clean['tulip'] + ty_probas['tulip'])/2
```

```python
## Choose max of the two
final_probas['daisy'] = np.maximum(scott_probas_clean['daisy'], ty_probas['daisy'])
final_probas['dandelion'] = np.maximum(scott_probas_clean['dandelion'], ty_probas['dandelion'])
final_probas['rose'] = np.maximum(scott_probas_clean['rose'], ty_probas['rose'])
final_probas['sunflower'] = np.maximum(scott_probas_clean['sunflower'], ty_probas['sunflower'])
final_probas['tulip'] = np.maximum(scott_probas_clean['tulip'], ty_probas['tulip'])
```

```python
probas = np.array(final_probas[['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']])

CLASSES = ['daisy', 'dandelion', 'rose', 'sunflower', 'tulip']
final_df = pd.DataFrame({'ID': final_probas['ID']})
final_df['Prediction'] = [CLASSES[i] for i in probas.argmax(axis=1)]
```

```python
final_df.to_csv('final_submission2.csv', index=False)
```
